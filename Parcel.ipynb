```python
import pandas as pd
import geopandas as geo
```

```python
#Read in Files
xls = pd.ExcelFile(r'Resources/Parcel/ParcelDB_20201118_SMZ_REDEV.xlsx')
xls2 = pd.ExcelFile(r'FullMont.xlsx')

Mont_Cont = pd.read_excel(xls, 'Mont_ET_Parcels')


Mont2020 = pd.read_excel(xls2, '2020')
Mont2030 = pd.read_excel(xls2, '2030')
Mont2040 = pd.read_excel(xls2, '2040')
Mont2050 = pd.read_excel(xls2, '2050')
Mont2060 = pd.read_excel(xls2, '2060')
Mont2070 = pd.read_excel(xls2, '2070')
Mont2080 = pd.read_excel(xls2, '2080')
Mont2090 = pd.read_excel(xls2, '2090')
Mont2100 = pd.read_excel(xls2, '2100')

MontSHP = geo.read_file("Montopolis/Montopolis.shp")
print(Mont2020.columns)
```

```python
DecadeArrayBest = [['2020 Best Type'],['2030 Best Type'],['2040 Best Type'],['2050 Best Type'],
                   ['2060 Best Type'],['2070 Best Type'],['2080 Best Type'],['2090 Best Type'],['2100 Best Type']]
DecadeArrayWorst = [['2020 Worst Type'],['2030 Worst Type'],['2040 Worst Type'],['2050 Worst Type'],
                    ['2060 Worst Type'],['2070 Worst Type'],['2080 Worst Type'],['2090 Worst Type'],['2100 Worst Type']]

DecadeArrayEBest = [['2020 EBest Type'],['2030 EBest Type'],['2040 EBest Type'],['2050 EBest Type'],
                    ['2060 EBest Type'],['2070 EBest Type'],['2080 EBest Type'],['2090 EBest Type'],['2100 EBest Type']]

DecadeArrayEWorst = [['2020 EWorst Type'],['2030EWorst Type'],['2040EWorst Type'],['2050EWorst Type'],['2060EWorst Type']
                     ,['2070 EWorst Type'],['2080 EWorst Type'],['2090 EWorst Type'],['2100 EWorst Type']]

decade = 2020
for ele in range (len(DecadeArrayBest)): 
    for index, row in Mont2020.iterrows():
        #print(row['RedevYear'])
        #print(int(decade))
        if(index == 0):
            print(decade,(int(row['RedevYear'])))
            print(int(decade) <= int(row['RedevYear']))
        if  int(decade) <= row['RedevYear']:
            DecadeArrayBest[ele].append(row['Current'])
            DecadeArrayWorst[ele].append(row['Current']) 
            
            DecadeArrayEBest[ele].append(row['Energy Current2020'])
            DecadeArrayEWorst[ele].append(row['Energy Current2020'])
            
        else:
            DecadeArrayBest[ele].append(row['Best Building'])
            DecadeArrayEBest[ele].append(row['Energy Best2020'])
            
            DecadeArrayWorst[ele].append(row['Best Building'])
            DecadeArrayEWorst[ele].append(row['Energy Worst2020'])
                  
    decade+=10
for ele in(DecadeArrayEBest):
    print(ele[0:7])
for ele in DecadeArrayBest:
    Mont2020.insert(len(Mont2020.columns),ele[0],ele[1:])
for ele in DecadeArrayWorst:
    Mont2020.insert(len(Mont2020.columns),ele[0],ele[1:])

for ele in DecadeArrayEBest:
    Mont2020.insert(len(Mont2020.columns),ele[0],ele[1:])
for ele in DecadeArrayEWorst:
    Mont2020.insert(len(Mont2020.columns),ele[0],ele[1:])
```

```python
def abc(x):
    x_as_string = str(x)
    return('Mont.'+x_as_string)
```

```python
def bcd(x):
    x_as_string= str(x)
    return('0'+x_as_string)
```

```python
MontBinder = Mont_Cont.loc[:,['OBJECTID','GEO_ID']]
```

```python
MontBinder
```

```python
MontBinder['OBJECTID'] = MontBinder['OBJECTID'].apply(abc)
MontBinder['GEO_ID'] = MontBinder['GEO_ID'].apply(bcd)
```

```python
MontBinder
result = pd.merge(Mont2020,MontBinder,left_on='ID',how='left',right_on='OBJECTID')
```
