```python
import pandas as pd
import plotly.graph_objects as go
from plotly.offline import plot,iplot, init_notebook_mode
def ImportCSV():

    xls = pd.ExcelFile(r'A1b Scenario 2.xlsx')
    Ref0 = pd.read_excel(xls, 'Ref0')
    Ref50 = pd.read_excel(xls, 'Ref50')
    Ref100 = pd.read_excel(xls, 'Ref100')
    
    City_Array = [Ref0, Ref50, Ref100]
    return City_Array
San_Antonio = ImportCSV()
df1= San_Antonio[0]
```

```python
df1
```

```python
spacex = df1.loc[:,'Year-0-A1b']
spacey = df1.loc[:,'Ref %']
spacez = df1.loc[:,'Cooling (Wh)']
```

```python

plot_scatter = go.Scatter3d(
    x = spacex,
    y = spacey,
    z = spacez,
    mode = 'markers',
    marker=dict(
    size = 2,
    color = spacez,
    colorbar =dict(title='Cooling Demand'),
    colorscale ='matter'
    )
)
plotData = [plot_scatter]
layout = go .Layout(
    margin = dict(
    l = 0,
    r = 0,
    b = 0,
    t = 0
    )
)
fig = go.Figure(data=plotData, layout=layout,)
iplot(fig,filename='Interactive_3dScatter.html')
```
